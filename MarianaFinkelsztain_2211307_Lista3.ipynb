3) Escolha uma coluna com valores numéricos e apague todas as linhas que possuam valor nulo


```python
#Tirando as linhas que possuem valor Na
nova_data=data.dropna()
display(nova_data)

#Como não tem nenhum valor Na, tirando as linhas que possuem valor 0 
filtro= data['ENTRADA']>0
nova_data=data[filtro]
display(nova_data)
print(nova_data["ENTRADA"])
```

4) Calcule a média desta mesma coluna

```python
#A média dessa coluna: 
media= data["ENTRADA"].mean()
print("Média: %.2f"%media)

#A média da nova coluna (considerando os 0 eliminados)
nova_media=nova_data["ENTRADA"].mean()
print("Nova Média: %.2f"%nova_media)
```

5) Escolha um critério para eliminar as linhas que possuam valores muito distantes da média e crie um novo Dataframe.

```python
#Considerando a média = 32.26
#Valores > 62 ou <2 são considerados muito distantes (+-30)

filtro_novo=abs(data['ENTRADA']-media)<=30
data2=data[filtro_novo]
display(data2)
```

6) Crie um novo Dataframe que contenha apenas 5 colunas do conjunto anterior.

```python
#Apenas 5 colunas
data3=pd.concat([data['RENDA'],data['VBEM'],data['VPARC'],data['IDADE'],data["RESMS"]],axis=1)
display(data3)
```

7) Escolha duas colunas numéricas e crie uma nova coluna que siga a regra: (Coluna A - Coluna B)*número aleatório (entre 0 e 1).

```python
#Escolhendo as Colunas RENDA e VBEM
import random as rd

n=rd.uniform(0,1)
print(n)

data['NOVA_COLUNA']=((data['RENDA']-data['VBEM'])*n)
display(data)
```

8) Embaralhe aleatoriamente o dataframe (índices e valores)

```python
import numpy as np
data4=data.reindex(np.random.permutation(data.index))
np.random.shuffle(data4.values)
display(data4)
```

9) Ordene o Dataframe seguindo um critério definido por você (Explique o critério)


```python
#Critério: Ordem crescente de idade
data=data.sort_values(by='IDADE',ascending=True)
print(f"A menor idade é: {data['IDADE'].iloc[0]}")
print(f"A maior idade é:{data['IDADE'].iloc[-1]}" )

display(data)
```

10) Escolha uma coluna numérica e um coluna com strings e utilize a função Groupby para mostrar os totais da  coluna


```python
#como não existe nenhuma coluna com strings, vou criar uma 

data['Strings']=('Mariana')
display(data)

#outra maneira é transformando uma das colunas em str
data['RESMS']=data['RESMS'].astype(str)

groupnum=data.groupby('IDADE')[['IDADE']].count()
groupstr=data.groupby('Strings')[['Strings']].count()
groupstr2=data.groupby('RESMS')[['RESMS']].count()

display(groupnum)
display(groupstr)
display(groupstr2)
```

11) Utilize a função MAPS para fazer algum tipo de substituição no Dataframe (explique)

```python
#Critério: se o Target for 1, substitui por UM 
# se o Target for 0, substitui por ZERO

data['target']=data['target'].map({1:"UM",0:"ZERO"})

display(data)
```
